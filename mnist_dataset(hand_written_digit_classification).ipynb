### The network architecture

```python
#2nd step : Network formation

from tensorflow.keras import models
from tensorflow.keras import layers

network = models.Sequential()

network.add(layers.Dense(512, activation='relu', input_shape=(28 * 28,)))
network.add(layers.Dense(10, activation='softmax'))

#first layer is input which has 784 (28 * 28) nodes then second layer has 512 \nodes and last layer has 10 nodes.
#last node is output and all middle layers are hidden layers.
#512 nodes are densely connected with 784 values and in middle there is activation function which converts linearity into non-linearity 
```

```python
#3rd step :  Compilation step
#compilation requires 3 things : loss function, optimizer, Metrics to monitor during training and testing
#Loss function : How the network will be able to measure its performance on the training data.
#Optimizer(backward propogation) : —The mechanism through which the network will update itself based on the data it sees and its loss function.
#Metrics to monitor during training and testing : Here, we’ll only care about accuracy (the fraction of the images that were correctly classified).
#To measure performance, we pass metrics.Matrics are of 3 types       Accuracy , precision , recall.
#Jis neuron p probability ki value zyada hogi wo usi class se belong krega.
```

```python
#For overall correctness : accuracy =  #correct / #total    (formula): ch:4
#Preision = TP / (TP + FP)
#Recall = TP / (TP + FN)          where TP = true positive, FP= False positive,  FN= False negative
```

```python
#Loss function :categorical_crossentropy(we use this loss function when classes are more than 2)
```

```python
network.compile(optimizer='rmsprop', loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
#In sequntial model, Dense method always take matrix in vector form so we need to preprocess it.
#preproess: before training, reshape data in that shape that network/model is expecting
#we should train matrix in same dimension that model is expecting
#We do scaling because there is large variation in data
#Scaled down : scale data so that data is in range of 0 and 1.
#For scaling, in data , divide every value by the largest value in data.
#In scaling there is no loss.
train_images[4]/255
```

```python
#here we are reshaping 3D matrix into 2Dmatrix.
train_images = train_images.reshape((60000, 28 * 28))
train_images = train_images.astype('float32') / 255
test_images = test_images.reshape((10000, 28 * 28))
test_images = test_images.astype('float32') / 255
```

```python
train_images
```

```python
train_images[0]
```

```python
#there are 2 types of data: Nominal and ordinal
#Nominal data: data on which we cannot apply quantative operations.
#Ordinal data: data on which we can apply quantative operations(like comparison).
#We have to change our labels(0-9) from ordinal to nominal otherwise model will also learn those labels and start performing quantitative operations on it like comparision which is wrong/dangerous fo model.
```

```python
from tensorflow.keras.utils import to_categorical
train_labels = to_categorical(train_labels)
test_labels = to_categorical(test_labels)
#to_categorical is used to convert class into matrix.
```

```python
train_labels[5]
```

```python
#we run epochs to increase accuracy.
network.fit(train_images, train_labels, epochs=5, batch_size=128)
```

```python
#if we want to change epoch like we want to run 20 epoch so output will return 25 epoch output because 5 are previously done
#to remove previous 5 re-initialize the model.(line after network architecture heading)
#if we run epochs large number of time then the accuracy will not be reliable because model is training on similar image multiple number of times.
#and also it will start learning some irrelevant features.
```

```python
#For train -----use fit
#For test ------use evaluate
test_loss, test_acc = network.evaluate(test_images, test_labels)
print('test_acc:', test_acc)
```
